```python
import pandas as pd
import matplotlib.pyplot as plt

# Чтение данных из файла Excel
data = pd.read_excel('E:\Desktop\A-B-test\sm8\AB_Test_Results.xlsx')

# Группировка данных по вариантам и вычисление среднего дохода для каждого варианта
variant_data = data.groupby('VARIANT_NAME')['REVENUE'].mean()

# Вывод результатов
print(variant_data)

# Построение графика
plt.bar(variant_data.index, variant_data.values)
plt.xlabel('Вариант')
plt.ylabel('Средний доход')
plt.title('Средний доход по вариантам')
plt.show()

plt.boxplot([data[data['VARIANT_NAME'] == 'control']['REVENUE'], data[data['VARIANT_NAME'] == 'variant']['REVENUE']], labels=['Control', 'Variant'])
plt.xlabel('Вариант')
plt.ylabel('Доход')
plt.title('Распределение дохода по вариантам')
plt.show()


```

Результаты эксперимента показывают, что текущие изменения, реализованные в варианте, не привели к увеличению дохода. Это может потребовать пересмотра идей и стратегий, использованных для варианта эксперимента.
Для достижения цели увеличения дохода рекомендуется провести дальнейшие исследования и анализ, чтобы выявить причины низкого дохода и разработать новые гипотезы для будущих экспериментов.
Рекомендуется использовать более длительный период и более крупную выборку пользователей для повторного проведения эксперимента. Это может помочь получить более надежные результаты и выявить потенциальные эффекты изменений.
Согласно анализу данных, полученных из эксперимента, средний доход для группы "control" составляет примерно 0.129013, в то время как средний доход для группы "variant" составляет примерно 0.070070.
Эти результаты указывают на то, что группа "control" имеет более высокий средний доход по сравнению с группой "variant". Однако для более точной интерпретации результатов необходимо провести статистический анализ и определить, является ли разница между группами статистически значимой.
