```python
import pandas as pd
import matplotlib.pyplot as plt

#     Excel
data = pd.read_excel('E:\Desktop\A-B-test\sm8\AB_Test_Results.xlsx')

#           
variant_data = data.groupby('VARIANT_NAME')['REVENUE'].mean()

#  
print(variant_data)

#  
plt.bar(variant_data.index, variant_data.values)
plt.xlabel('')
plt.ylabel(' ')
plt.title('   ')
plt.show()

plt.boxplot([data[data['VARIANT_NAME'] == 'control']['REVENUE'], data[data['VARIANT_NAME'] == 'variant']['REVENUE']], labels=['Control', 'Variant'])
plt.xlabel('')
plt.ylabel('')
plt.title('   ')
plt.show()


```
